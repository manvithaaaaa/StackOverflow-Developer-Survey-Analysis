```python
df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['ConvertedCompYearly'] = pd.to_numeric(
    df['ConvertedCompYearly'], errors='coerce')
```

```python
df['YearsCode'].fillna(df['YearsCode'].median(), inplace=True)
df['YearsCodePro'].fillna(df['YearsCodePro'].median(), inplace=True)
```

```python
df['Employment'].fillna(df['Employment'].mode()[0], inplace=True)
df['RemoteWork'].fillna(df['RemoteWork'].mode()[0], inplace=True)
df['EdLevel'].fillna(df['EdLevel'].mode()[0], inplace=True)
df['Country'].fillna(df['Country'].mode()[0], inplace=True)
```
